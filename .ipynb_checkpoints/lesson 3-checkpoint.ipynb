```python
import pandas as pd
import numpy as np
import implicit
import dask.dataframe as dd

import warnings
warnings.filterwarnings('ignore')
```

```python
from implicit.als import AlternatingLeastSquares
from scipy.sparse import csr_matrix
```

```python
data = pd.read_csv('data/retail_train.csv')
# data = data.astype(np.float32)
data
```

```python
data.__sizeof__()
```

```python
for column in data.columns:
    if np.float16(data[column].max()) != np.inf:
        data[column] = data[column].astype(np.float16)
    elif np.float32(data[column].max()) != np.inf:
        data[column] = data[column].astype(np.float32)
```

```python
data.__sizeof__()
```

```python
# from dask.dataframe import methods

# data = data.categorize(columns=['PRODUCT_ID',])
# methods.is_categorical_dtype(data['PRODUCT_ID'])
```

```python
data_user_item = pd.pivot_table(data[:100000], index='item_id', columns='basket_id', 
                                  values='quantity', aggfunc='sum', fill_value=0)
data_user_item
```

```python
model = AlternatingLeastSquares(factors=100,
                                 regularization=0.01,
                                 iterations=15,
                                 calculate_training_loss=True,
                                 num_threads=4)
```

```python
model.fit(csr_matrix(data_user_item))
```

```python
list_id = pd.DataFrame(np.array(data_user_item.columns).T, columns=['id',])
list_id
```

```python
np.array([list(i) for i in model.recommend(userid=0, user_items=csr_matrix(data_user_item), N=5)])[:,0]
```

```python
model.user_norms.shape, model.item_norms.shape
```

```python
def recomended(x_list, k=3):
    return [np.array([list(i) for i in model.recommend(userid=x, user_items=csr_matrix(data_user_item), N=k)])[:,0] for x in x_list]
```

```python
x = [0, 1, 2, 3]
recomended(x)
```

```python
x = [0, 1, 2, 3]
recomended(x, 5)
```
